###### Description: This data set was made available for the Physiological Data Modeling Contest at ICML 2004. The data was collected from subjects using BodyMedia wearable body monitors while performing their usual activities. These monitors record acceleration, heat flux, galvanic skin response, skin temperature, and near-body temperature. The training data set includes several sessions for each of multiple subjects, with measurements stored each minute during a session. The test data set includes further sessions from the same subjects, as well as sessions recording measurements from new subjects who did not feature in the training data. Each record in the data includes an annotation code giving information about the kind of activity that the subject was performing at that time. Participants in the competition were asked to train classifiers to apply two of these annotation codes to the test data, and also to train a classifier to identify subjects as men or women (this information is given in the training data sequences).

###### Métodos úteis na customização dos gráficos e exibição dos dados

```python
def remove_border(axes=None, top=False, right=False, left=True, bottom=True):

    ax = axes or plt.gca()
    ax.spines['top'].set_visible(top)
    ax.spines['right'].set_visible(right)
    ax.spines['left'].set_visible(left)
    ax.spines['bottom'].set_visible(bottom)
    
    #turn off all ticks
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')
    
    #now re-enable visibles
    if top:
        ax.xaxis.tick_top()
    if bottom:
        ax.xaxis.tick_bottom()
    if left:
        ax.yaxis.tick_left()
    if right:
        ax.yaxis.tick_right()

#retorna dados de um usuario durante uma sessao
def getUserSubject(userID, sessionID, subject):
    return np.array(dados[(dados.userID == userID) & (dados.sessionID == sessionID)][subject])

#plota a serie temporal de um usuario durante uma sessao
def plotSubjectSeriesTime(userID,sessionID,subject, movavg=False, color='green',n=0,alpha=1):
    plt.figure(figsize=(12,6))
    data=getUserSubject(userID,sessionID,subject)
    if(movavg==False):
        plt.plot(pd.Series(data),alpha=alpha,color=color)
    else:
        plt.plot(pl.movavg(data,n),color=color,alpha=alpha)
    plt.title(subject)
    plt.xlabel('minutes')
    remove_border()
    plt.show()
```

```python
# dicionário associando o userID de cada usuário (key) com uma lista de sessionID (values)

userIDs = np.array(dados['userID'])
userIDs = np.unique(userIDs)

userSession = {}
[userSession.setdefault(key, []) for key in userIDs]

for userID in userIDs:
    sessionIDs = []
    [sessionIDs.append(i) for i in dados[dados.userID == userID]['sessionID']]
    sessionIDs = np.unique(sessionIDs)
    [userSession[userID].append(i) for i in sessionIDs]
```
